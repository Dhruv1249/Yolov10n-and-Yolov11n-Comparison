```python
import pandas as pd
import matplotlib.pyplot as plt

results_10n = pd.read_csv(r'/content/drive/MyDrive/yolov10nresults/content/runs/detect/train/yolov10n-res.csv')
results_11n = pd.read_csv(r'/content/drive/MyDrive/yolo11nResults/content/runs/detect/train/yolov11n.csv')

epochs_10n = results_10n['epoch']
precision_10n = results_10n['metrics/precision(B)']

epochs_11n = results_11n['epoch']
precision_11n = results_11n['metrics/precision(B)']

plt.figure(figsize=(10, 5))
plt.plot(epochs_10n, precision_10n, label='YOLOv10n', color='blue', linestyle='--', marker='o')
plt.plot(epochs_11n, precision_11n, label='YOLOv11n', color='green', linestyle='-', marker='s')
plt.xlabel('Epochs')
plt.ylabel('Precision (B)')
plt.title('YOLOv10n vs YOLOv11n - Precision over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

epochs_10n = results_10n['epoch']
recall_10n = results_10n['metrics/recall(B)']

epochs_11n = results_11n['epoch']
recall_11n = results_11n['metrics/recall(B)']

plt.figure(figsize=(10, 5))
plt.plot(epochs_10n, recall_10n, label='YOLOv10n', color='orange', linestyle='--', marker='o')
plt.plot(epochs_11n, recall_11n, label='YOLOv11n', color='red', linestyle='-', marker='s')
plt.xlabel('Epochs')
plt.ylabel('Recall (B)')
plt.title('YOLOv10n vs YOLOv11n - Recall over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
epochs_10n = results_10n['epoch']
map50_10n = results_10n['metrics/mAP50(B)']

epochs_11n = results_11n['epoch']
map50_11n = results_11n['metrics/mAP50(B)']

plt.figure(figsize=(10, 5))
plt.plot(epochs_10n, map50_10n, label='YOLOv10n', color='green', linestyle='--', marker='o')
plt.plot(epochs_11n, map50_11n, label='YOLOv11n', color='blue', linestyle='-', marker='s')
plt.xlabel('Epochs')
plt.ylabel('mAP@0.5 (B)')
plt.title('YOLOv10n vs YOLOv11n - mAP@0.5 over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

epochs_10n = results_10n['epoch']
map5095_10n = results_10n['metrics/mAP50-95(B)']

epochs_11n = results_11n['epoch']
map5095_11n = results_11n['metrics/mAP50-95(B)']

plt.figure(figsize=(10, 5))
plt.plot(epochs_10n, map5095_10n, label='YOLOv10n', color='purple', linestyle='--', marker='o')
plt.plot(epochs_11n, map5095_11n, label='YOLOv11n', color='red', linestyle='-', marker='s')
plt.xlabel('Epochs')
plt.ylabel('mAP@0.5:0.95 (B)')
plt.title('YOLOv10n vs YOLOv11n - mAP@0.5:0.95 over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
